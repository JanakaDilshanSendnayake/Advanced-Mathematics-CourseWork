```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
import math
```

```python
def plot_function(x_values, y_values, label, title, x_label, y_label):
    plt.plot(x_values, y_values, label=label)
    plt.title(title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
# Define the function
def f(x):
    return x * np.sin(x/4)


# Generate x values
x_values = np.linspace(-6*np.pi, 3*np.pi, 1000)

# Generate y values
y_values = f(x_values)

# Plot the function
plot_function(x_values, y_values,'$f(x) = xsin(x/4)$','Plot of the function $f(x) = xsin(x/4)$','x','$sin(x/4)$')
```

**The Taylor series for $sin(x)$ centerd at $x= (\frac{\pi}{4})$ is given by :**

$sin(x) \approx \sum_{n=0}^{\infty} \frac{\sin^{(n)}(\frac{\pi}{4})}{n!}(x - \frac{\pi}{4})^{n} $

$\therefore n^{th}$ term of Taylor series:
$T_n(x) = \frac{\sin^{(n)}(a)}{n!}(x - a)^{n}
 $

```python
# Get the taylor series expansion for sin(x) (An approximation for first 50 terms)
x=sp.symbols('x')
f=sp.sin(x)
a = sp.pi / 4 
taylor_series_50 = f.subs(x, a) 
num_terms = 50

for index in range(1, num_terms):
    taylor_series_50 += sp.diff(f, x, index).subs(x, a)*(x-a)**index / sp.factorial(index)
taylor_series_50
```

```python
taylor_series_50_numerical=sp.lambdify(x,taylor_series_50,'numpy')
values=taylor_series_50_numerical(x_values)
plot_function(x_values, values,'','Plot of the Tylor series approximation of $f(x) = sin(x/4)$ for 1st fifty terms','x','$sin(x/4)$ \approx')
```

```python
approximation = taylor_series_50_numerical(np.pi / 4)*np.pi / 2
actual_value = (np.pi / 2) * np.sin(np.pi / 4)

# Calculate absolute error
absolute_error = np.abs(actual_value-approximation )

# Calculate percentage error
percentage_error = (absolute_error / np.abs(actual_value)) * 100

print(f"Actual Value: {actual_value}")
print(f"Approximation: {approximation}")
print(f"Absolute Error: {absolute_error}")
print(f"Percentage Error: {percentage_error}%")
```

Let's simplify taylor series expansion of $\sin(x)$ centered at $x=\pi /4$ by taking out the first term from sigma: $\sqrt{2}/2 + \sum_{n=1}^{\infty} \frac{\sin^{(n)}(\pi/4)}{n!}(x - \pi/4)^{n} $

When $x=\pi /4$ , expression inside sigma of above simplified expression is always zero.  $\therefore$ The Taylor series of $\sin(x)$ centered at $\pi/4$ always returns $\sqrt{2}/2$ when $x=\pi/4$. $\therefore$ in this case $x=\pi/4$ isn't a suitable value to demonstrate the deviation between $\sin(x)$ and it's taylor series centered at $x=\pi /4$.
