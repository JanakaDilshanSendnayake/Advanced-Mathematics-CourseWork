```python
import numpy as np
import matplotlib.pyplot as plt
```

A function to plot a graph for a given set of x and y values

```python
def plot_function(x_values, y_values, label, title, x_label, y_label):
    plt.plot(x_values, y_values, label=label)
    plt.title(title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.legend()
    plt.grid(True)
    plt.show()
```

#### a)

$f(x) = \frac{1}{1 + e^{-x}}$

```python
def logistic_function(x):
    return 1 / (1 + np.exp(-x))

x_values = np.linspace(-10, 10, 100000)
y_values = logistic_function(x_values)

plot_function(x_values, y_values, 'Logistic Function', 'Logistic Function Plot', 'x', 'f(x)')
```

#### b)

Derivative of a logistic function is given by : $f'(x) = \frac{e^{x}}{(1 + e^{x})^2} $

```python
def logistic_derivative(x):
    u = 1 + np.exp(x)
    return np.exp(x) / u**2
derivative_of_logistic_function_y = logistic_derivative(x_values)
plot_function(x_values,derivative_of_logistic_function_y,'Derivative of Logistic Function', 'Derivative of Logistic Function Plot', 'x', 'f\'(x)')
```

```python
# Derivative of logistic function purely using Numpy

derivative_of_logistic_function = np.gradient(y_values, x_values)
plot_function(x_values,derivative_of_logistic_function,'Derivative of Logistic Function', 'Derivative of Logistic Function Plot', 'x', 'f\'(x)')
```

#### c)

Different ranges for $x$ were used to highlight their special characterisitics

$tan(sin(3x)^2)$

```python
def function_a(x):
    return np.tan((np.sin(3 * x))**2)

a_y_values = function_a(x_values)
plot_function(x_values, a_y_values, '$tan(sin(3x)^2)$', '$tan(sin(3x)^2)$ Plot', 'x', 'f(x)')
```

$-x^4+2x^2-{1/2}x+10$

```python
def function_b(x):
    return -x**4 + 2 * x**2 - 0.5 * x + 10
x_values = np.linspace(-2, 2, 100000)
b_y_values = function_b(x_values)
plot_function(x_values, b_y_values, '$-x^4+2x^2-{1/2}x+10$', '$-x^4+2x^2-{1/2}x+10$ Plot', 'x', 'f(x)')
```

$e^{-xlnx^2}$

```python
def function_c(x):
    return np.exp(-x * np.log(x**2))
x_values = np.linspace(-2, 2, 100000)
c_y_values = function_c(x_values)
plot_function(x_values, c_y_values, '$e^{-xlnx^2}$', '$e^{-xlnx^2}$ Plot', 'x', 'f(x)')
```

$x^2sin(sin(3x))-3cos(cos(2x-\pi/3))$

```python
def function_d(x):
    return x**2 * np.sin(np.sin(3 * x)) - 3 * np.cos(np.cos(2 * x - np.pi/3))
x_values = np.linspace(-15, 15, 100000)
d_y_values = function_d(x_values)
plot_function(x_values, d_y_values, '$x^2sin(sin(3x))-3cos(cos(2x-\pi/3))$', '$x^2sin(sin(3x))-3cos(cos(2x-\pi/3))$ Plot', 'x', 'f(x)')
```
