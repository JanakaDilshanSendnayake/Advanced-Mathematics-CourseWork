```python
import numpy as np
import matplotlib.pyplot as plt
```

#### A common function to plot Sine wave of a given frequency sampled at a given sampling rate and it's maginitude spectrum

```python
def plot_aliasing_demo(frequency, sampling_rate, duration):
    # Generate a sine wave with the given frequency
    t = np.arange(0, duration, 1/sampling_rate)
    signal = np.sin(2 * np.pi * frequency * t)

    # Plot the original signal and magnitude spectrum
    plt.figure(figsize=(16, 8))

    plt.subplot(2, 2, 1)
    plt.plot(t, signal)
    plt.title(f'Original Signal: {frequency} Hz, Sampling Rate: {sampling_rate} Hz')
    plt.xlabel('Time (s)')
    plt.ylabel('Amplitude')

    # Perform FFT on the signal
    fft_result = np.fft.fft(signal)
    fft_freq = np.fft.fftfreq(len(fft_result), d=1/sampling_rate)

    # Plot the magnitude spectrum
    plt.subplot(2, 2, 2)
    plt.plot(fft_freq, np.abs(fft_result))
    plt.title('Magnitude Spectrum')
    plt.xlabel('Frequency (Hz)')
    plt.ylabel('Magnitude')
    plt.xlim(0, sampling_rate/2)  # Display only positive frequencies

    plt.tight_layout()
    plt.show()
```

# Case 1: Sampling Rate Above Nyquist Rate

### Example 1
In this example, I've chosen a sine wave with a frequency of 20 Hz and a sampling rate of 50 Hz. The Nyquist theorem states that the sampling rate should be at least twice the frequency of the signal to avoid aliasing. In this case, the Nyquist frequency is $( \frac{{\text{{sampling rate}}}}{2} = \frac{50}{2} = 25 )$ Hz. Since the frequency of our sine wave (20 Hz) is lower than the Nyquist frequency, aliasing won't occur.


```python
frequency = 20  # Frequency of the sine wave (in Hz)

# Case 1: Original Sampling Rate is below Nyquist Rate
sampling_rate = 50  # Sampling rate (in Hz)
duration = 1  # Duration of the signal (in seconds)
plot_aliasing_demo(frequency, sampling_rate, duration)
```

# Summary

### When Nyquist Theorem is Satisfied:
1. The magnitude spectrum correctly identifies the true frequency of the signal as the highest peak, aligning with the original frequency of 20 Hz.

### When Nyquist Theorem is Violated
1. Undersampling results in aliasing, where the true frequency is misrepresented in the magnitude spectrum.
2. The highest peak in the magnitude spectrum points to a frequency that is different from the actual frequency of the sine wave.
3. Aliasing introduces errors in the frequency domain representation, leading to a distorted understanding of the signal's frequency content.
